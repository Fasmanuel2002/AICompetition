```python
import pandas as pd

model_errors = "model_errors.csv"
predictions_model = "oof_predictions_with_meta.csv"
```

```python
df_errors = pd.read_csv(model_errors)
total_size = len(df_errors)

pred_label = df_errors["pred_label"].unique()
for classes in pred_label:
    df_loop = df_errors[df_errors["pred_label"] == classes]
    
    total_size_per_group = len(df_loop)
    
    print(f"{(total_size_per_group / total_size) * 100:.2f}%:  {classes}" )

```

```python
df_model_predictions = pd.read_csv(predictions_model)
df_model_predictions.head()

number_total_false = (~df_model_predictions["correct"]).sum()
number_total_True = (df_model_predictions["correct"] == True).sum()
total_rows = df_model_predictions.index.size
print(f"Model Errors of model prediction {(number_total_false / total_rows) * 100:.2f}%")
print(f"True predictions of model prediction {(number_total_True / total_rows) * 100:.2f}%")
```
